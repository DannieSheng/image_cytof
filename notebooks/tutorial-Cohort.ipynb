## Process a cohort
Process all images in a specified cohort by going over all previously introduced steps, save results, as well as a file containes scaling parameters calculated from all normalized features from the batch.

There are 2 initialization methods for **CytofCohort**:
1. Initialize with a *df_cohort*
    - Use class method *batch_process* to perform initializing *CytofImage* class instances and go through all processing steps
        - Meanwhile, scale all features use class method *scale_feature*
    - The class attribute *df_cohort* got updated upon the completion of *batch_process*
    - Create and attach the class attribute *cytof_images* upon the completion
2. Initialize with a dictionary of CytofImages: *cytof_images*
    - Use class method *batch_process_feature*
        - Meanwhile, scale all features use class method *scale_feature*
    - ~~Save each individua *CytofImage* instance object~~
    - Create and attach class attribute *df_cohort* upon the completion of *batch_process_feature*
